```python
import pandas as pd
```

## Read data and get unique specialties

```python
# Read Data
data = pd.read_stata("../data/hdf_doctor_info_original.dta")
data
```

```python
# Get unique specialties
doc_keshi = list(data["doc_keshi"].unique())
```

```python
len(doc_keshi)
```

```python
# remove empty string
doc_keshi = list(filter(None,doc_keshi))
```

```python
len(doc_keshi)
```

## Using AI to classify

```python
#Define the prompt
PROMPT = """As an AI trained with a vast medical knowledge base, I need you to classify various doctor specialties into two distinct groups: those primarily involved in the treatment of chronic diseases and those focused on non-chronic diseases. Chronic diseases are defined as conditions that persist for one year or more, necessitating ongoing medical attention or limiting activities of daily living, or both. Return '1' to indicate specialties that primarily deal with chronic diseases and '0' for those that do not. ONLY RETURN 0 or 1.

Doctor specialty:
"""
```

```python
# Define OpenAI Key
import os
os.environ["OPENAI_API_KEY"] = "YOUR KEY HERE"
```

```python
import openai


final_results = []
for i, sample in enumerate(doc_keshi):
    prompt = PROMPT + sample


    # call to openai
    response = openai.chat.completions.create(messages=[{"role": "user", "content": prompt}], model="gpt-4o")
    result = response.choices[0].message.content

    # append result to final list
    final_results.append(result)
    print(i, sample, result)
```

```python
# Save result to dict
result_dict = {}
for i, sample in enumerate(doc_keshi):
    result_dict[sample] = final_results[i]
```

```python
# get result for each row
is_chronic_column = []
for i, row in data.iterrows():
    if row["doc_keshi"] != "":
        is_chronic_column.append(result_dict[row["doc_keshi"]])
    else:
        is_chronic_column.append(None)
```

```python
len(is_chronic_column)
```

```python
len(data)
```

```python
# add new column
data["is_chronic"] = is_chronic_column
```

```python
# save new file
data.to_csv("../output/hdf_doctor_specialty_llm.csv", encoding="utf-8", index=False)
```

```python
# Save result to dict
result_dict = {}
for i, sample in enumerate(doc_keshi):
    result_dict[sample] = final_results[i]
```

```python
# get result for each row
specialty_group_column = []
for i, row in data.iterrows():
    if row["doc_keshi"] != "":
        specialty_group_column.append(result_dict[row["doc_keshi"]])
    else:
        specialty_group_column.append(None)
```
